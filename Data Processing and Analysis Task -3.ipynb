```python
#Problem Statement 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
    }

#step 1:find highest score
highest_score = max(employees.values())

#step 2:Find employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

#step 3:Display result
print("Top performers Eligible for Bonus:",",".join(top_performers), f"(Score:{highest_score})")
      
```

```python
# Problem Statement 2: Search Query Keyword Analysis

text = "Buy mobile phone buy phone online"
#step 1:convert text into lowercase
text= text.lower()

#step 2:remove punctuations
import string 
for ch in string.punctuation:
    text = text.replace(ch, "")
    
#step 3: split into words
words = text.split()

#step 4 :count freqency
freq ={}
for word in words:
    freq[word] = freq.get(word, 0) + 1
    
#step 5: show  words used more than once
result = {word: count for word,count in freq.items() if count >1}
print(result)
```

```python
# Problem Statement 3: Sensor Data Validation

sensor_readings =[3, 4, 7, 8 ,10, 12, 5]
#create empty list to store valid readings
valid_readings = []

#go through each reading with its  index(hour_
for hour, value in enumerate(sensor_readings):
#check if reading is even
    if value % 2 == 0:
    #store hour and reading value as a tuple
        valid_readings.append((hour, value))
        
print("Valid Sensor Readings (Hour, value):")
print(valid_readings)
    
    
```

```python
# Problem Statement 4: Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}
total = len(emails)

for email in emails:
    domain = email.split("@")[1]

    # count domains correctly
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# print percentages
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem Statement 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
#create dictionary to store count of each ID
id_count = {}

#count occurencies of each user ID
for uid in user_ids:
    if uid in id_count:
        id_count[uid] +=1
    else:
        id_count[uid] = 1
        
#display duplicate iDs
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} -> {count} times")
```
