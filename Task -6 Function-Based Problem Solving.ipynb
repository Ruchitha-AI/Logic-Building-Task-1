```python
# Problem Statement 1:Smart Parking Lot Management System

def smart_parking_managing_system(logs,capacity):
    #creating variable to store no.of parked cars
    current_parked_cars = 0
    peak = 0
    
    #loop through each log entry
    for log in logs:
        #if vehicle enters
        if log == "IN":
            current_parked_cars += 1
        #if vehicle exists
        elif log == "OUT" and current_parked_cars > 0:
            current_parked_cars -= 1
        #peak count
        if current_parked_cars > peak:
            peak = current_parked_cars
            
    #checking parking system
    if current_parked_cars > capacity:
        status = "Alert! Parking full "
    else:
        status = "Available"
    #displaying results
    print("Currently Parked vehicles:", current_parked_cars)
    print("Parking status:", status)

#sample input
capacity =  50
logs = ["IN", "IN", "IN", "OUT","IN", "OUT", "IN",]

# calling function
smart_parking_managing_system(logs,capacity)
    
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    time = distance * 5  # 1 km distance takes 5 minutes, basic speed assumption

    # Time based on traffic level
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 8
    elif traffic == "Low":
        time += 0

    #adjusting time based on weather  condition
    if weather == "Rainy":
        time += 5
    elif weather == "normal":
        time += 0
    
    #Displaying result
    print("Estimated Delivery Time:", time, "minutes")
    
#Sample Input
estimate_delivery_time(8, "High", "Rainy")



```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

def seat_occupancy(total_seats, booked_seats):
    # counts how many sests are booked
    booked = len(booked_seats)
    # calculate occupancy percentage
    occupancy = (booked / total_seats) * 100

    #show status based on occupancy
    if occupancy == 100:
        status = "Houseful"
    elif occupancy > 74:
        status = "Almost full"
    else:
        status = "Seats Available"

    #displaying occupancy percentage
    print("Occupancy:",int(occupancy), "%")
    # Showstatus displaying
    print("Show Status:", status)

#sample input
total_seats = 200
booked_seats = [1] * 150 

#calling fucntion
seat_occupancy(total_seats, booked_seats,)
                
    
```

```python
# Problem Statement 4:Cloud Server Load Classification System

def server_load_analyzing(CPU_Readings):
    #create variable to store total value of readings
    total = 0
    for reading in CPU_Readings:
        total += reading
    #calculate Average
    Average_CPU_load = total / len(CPU_Readings)
    
    #Evaluating Server status based on occupancy
    if  Average_CPU_load < 50:
        Server_Status = "Normal"
    elif 50 <= Average_CPU_load < 80:
        Server_Status = "Warning"
    else:
        Server_Status = "Critical"
        
    #Displaying output
    print("Average CPU Load:", int(Average_CPU_load),"%")
    print("Server Status:",  Server_Status)
    
#Sample input
CPU_Readings= [45, 60, 70, 85, 90]

#calling function
server_load_analyzing(CPU_Readings)


        
        
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor

def track_classroom_resources_usage(Usage_data):
    limit = 8

    #create variable to store overused resource
    overused = [ ]
    #check each resource
    for resource, hours in Usage_data.items():
        if hours > limit:
            overused.append(resource)
    #display results
    if overused:
        print("Overused Resources:",",".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

#Sample input
Usage_data ={
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

#calling function
track_classroom_resources_usage(Usage_data)
    
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller

def manage_registrations(capacity,registrations):
   #create variable to store Confirmed and Waitlist registrations
    confirmed = 0 
    waitlist = 0
    
    #Going through each user registering
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1
            
    #check status
    if confirmed == capacity:
        status = "closed"
    else:
        status = "Open"
        
    #displaying results
    print("Confirmed Registrations:",confirmed)
    print("Waitlisted Users:",waitlist)
    print("Registrations Status:", status)

#Sample input
manage_registrations(100,105)
    
    
```
