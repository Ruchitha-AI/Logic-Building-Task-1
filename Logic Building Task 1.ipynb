```python
#1. User Login Check

username = input ("Enter a username:")
password = input ("Enter a password:")

#checking user credentials
if username  == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2.Pass/Fail Analyzer

marks = [45, 78, 90, 33, 60]

#Initialize counters for pass and fail
pass_count = 0
fail_count = 0

#loop through each mark in the list
for mark in marks:
    if mark >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1
        
#Displaying final results
print(f"Total number of passed students: {pass_count}")
print(f"Total number of failed students: {fail_count}")
print(f"Total number of students: {len(marks)}")
```

```python
#3.Simple Data cleaner

#Input list containing uncleaned user names
names = [" Alice ","bob"," CHARLIE "]

#initialize an empty list to store cleaned names
cleaned_list = [ ]

#Loop through names
for name in names:
    cleaned_name = name.strip( ).lower( )  #strip() removes extra spaces,lower() converts to lowercase
    cleaned_list.append(cleaned_name)

#Displaying cleaned list
print("Cleaned list:", cleaned_list)
```

```python
#4.Message length Analyzer

messages = ["Hi","Welcome to the platform","OK"]

#Loop through each message
for msg in messages:
    length = len(msg)
    
    print("Messages:", msg)
    print("Length:", length)

#Checks if msg length is greater than 10 characters
    if length > 10:
        print("Flag: Message Longer than 10 characters")
    print()
    
```

```python
#5.Error Message Detector

logs = ["INFO","ERROR","WARNING","ERROR"]

#count how many times ERROR appears in the list
ERROR_COUNT = logs.count("ERROR")

print("Number of ERROR entries:",ERROR_COUNT)
```
