```python
#Problem 1: Unique Words in a Sentence

sentence = input("Enter sentence: ")
#split the sentence into a list of words
words_list = sentence.split() # split() breaks the string at every space
#Convert the list into a set
unique_words = set(words_list) # A set automatically  removes duplicate values
print("Unique words count:",len(unique_words))
print("Unique words:", unique_words)
```

```python
#Problem 2: Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# max() fucntion can be used on dictionary keys with a key function
highest_salary_employee = max(employees, key=employees.get) # The key function tells max() to compare based on salary values
print("Highest Salary:", highest_salary_employee, "-",employees[highest_salary_employee])
```

```python
#Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]
#Initialize max and min with the  frst element of the list
max_value = numbers[0]
min_value = numbers[0]
#loop through each number in the list
for num in  numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)

```

```python
#Problem 4: Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]
#Initialize a counter  to keep track of products  above 1000
count = 0
#Loop through each price in the list
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)

```

```python
#Problem 5: Calculate Attendance Percentage

attendance = [ "P", "P", "A", "P", "P" ]
#count the no.of "P" Present in the list
present_days = attendance.count("P")
#Calculate total number of days
total_days = len(attendance)
#calculate attendance percenatge 
attendance_percentage = (present_days / total_days )* 100
print("Attendance Percentage:", attendance_percentage)


```

```python
# Problem 7: Count Character Frequency

text = "pythonp"
#create an empty dictionary to store character count
char_count = {}
#loop through each cahr in the string
for char in text:
    if char in char_count: 
        char_count[char] += 1  # if char already in the dict increment its count
    else:
        char_count[char] = 1  # if char not in dict,add it with count 1
print(char_count)
```

```python
# Problem 9: Check if a Key Exists in a Dictionary

employees = {
    "Ravi" : 75000,
    "Anita" : 68000,
    "Kiran" : 72900
}
# specify the key you want to check
key_to_check = "Kiran"
# use in keyword to check  if the  key exists
if key_to_check  in employees:
    print("Employees exists")
else:
    print("Employee does not exist")
```

```python
# Problem 10: Calculate Average Marks

marks = [70, 75, 65, 80]
total_marks = sum(marks)
#calcualte total number of marks
number_of_marks = len(marks)
#calculate the average
average_marks = total_marks / number_of_marks
print("Average Marks:", average_marks)
```
